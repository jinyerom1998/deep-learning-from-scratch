# 3장 - 신경망
퍼셉트론의 단점: 가중치를 설정하는 작업을 사람이 해야 한다.<br>퍼셉트론의 단점을 해결한 것: 신경망

### 신경망의 예
신경망을 그림으로 나타내면 그림처럼 된다.<br>
![신경망의 예](그림자료들/chapter3-1.png)<br>
은닉층은 사람 눈에는 보이지 않는다.<br>
이 책에서는 입력층: 0층, 은닉층: 1층, 출력층: 2층이라 표현한다.<br>
위 그림의 신경망은 모두 3층으로 구성되어 있지만,<br>
가중치를 갖는 층은 2개 뿐이기 때문에 '2층 신경망'이라고도 한다.

### 퍼셉트론 복습
기존 학습했던 퍼셉트론 네트워크에서 편향 b를 추가하면 편향을 명시한 퍼셉트론이 된다.<br>
![편향을 명시한 퍼셉트론](그림자료들/chapter3-3.jpg)<br>
가중치가 b이고 입력이 1인 뉴런이 추가된 모습이다.<br>
이 퍼셉트론의 동작은<br>
x1, x2, 1이라는 3개의 신호가 뉴런에 *입력*되어,<br>
각 신호에 가중치를 곱한 후, 다음 뉴런에 전달된다.<br>
다음 뉴런에서는 이들 신호의 값을 더하여, 그 합이 0을 넘으면 1을 출력하고 그렇지 않으면 0을 출력한다.<br>
참고로, 편향의 입력 신호는 항상 1이기 때문에 그림에서는 해당 뉴런을 회색으로 채워 다른 뉴런과 구분했다.<br>
<br>
퍼셉트론의 식은 다음과 같이 간결하게 변환 가능하다.<br>
